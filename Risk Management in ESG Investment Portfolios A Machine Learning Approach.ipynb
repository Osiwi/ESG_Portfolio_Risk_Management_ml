```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from scipy.optimize import minimize
import xgboost as xgb
from sklearn.utils import resample

# Step 1: Load and preprocess the data

# DATA SOURCE / CONFIDENTIALITY NOTE
# The original dataset contains proprietary ESG time-series data and macroeconomic indicators.
# For data privacy, the raw file (ESG_STOCK_DATA.xlsx) is NOT included in this public repository,
# and the local machine path has been removed.
#
# To reproduce this analysis:
# 1. Prepare your own dataset with similar columns, e.g.:
#    - Date
#    - Company ESG_Score columns (e.g. CompanyA_ESG_Score, CompanyB_ESG_Score, ...)
#    - Macroeconomic indicators (Inflation Rate, Interest Rate, GDP Growth Rate)
# 2. Save it locally (not committed to GitHub) as: data/ESG_STOCK_DATA.xlsx
# 3. Update the path below if needed.
#
# The code below expects the file at: data/ESG_STOCK_DATA.xlsx

file_path = "data/ESG_STOCK_DATA.xlsx"  # relative path to your private/local copy
data = pd.read_excel(file_path)


# Convert 'Date' column to datetime
if 'Date' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'])

# Define ESG and macro columns
esg_cols = [col for col in data.columns if 'ESG_Score' in col]
macro_cols = ['Inflation Rate', 'Interest Rate', 'GDP Growth Rate']

# Calculate 30-day rolling volatility for ESG scores
for company in esg_cols:
    data[f'{company}_Volatility'] = data[company].rolling(window=30).std()

# Step 2: Define features and labels
X = data[esg_cols + macro_cols].fillna(0)
volatility_median = data[[f'{company}_Volatility' for company in esg_cols]].mean(axis=1).median()
y = (data[[f'{company}_Volatility' for company in esg_cols]].mean(axis=1) > volatility_median).astype(int)  # Binary classification: high/low risk

# Step 3: Balance the data using oversampling
data_resample = pd.concat([X, y], axis=1)
data_resample.columns = list(X.columns) + ['Risk']
df_majority = data_resample[data_resample.Risk == 0]
df_minority = data_resample[data_resample.Risk == 1]

df_minority_oversampled = resample(df_minority, replace=True, n_samples=len(df_majority), random_state=42)
data_balanced = pd.concat([df_majority, df_minority_oversampled])

X_balanced = data_balanced.drop('Risk', axis=1)
y_balanced = data_balanced['Risk']

# Step 4: Train-test split and scaling
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Function to plot confusion matrix heatmap
def plot_confusion_matrix_heatmap(model, X_test, y_test, model_name):
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(y_test, model.predict(X_test))
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Confusion Matrix: {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

# Step 5: MLPClassifier with GridSearchCV
param_grid_mlp = {
    'hidden_layer_sizes': [(64, 32, 16), (128, 64), (128, 64, 32)],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate_init': [0.001, 0.01],
    'max_iter': [1000],
    'activation': ['relu', 'tanh'],
    'solver': ['adam', 'lbfgs']
}

mlp_model = MLPClassifier(random_state=42)
grid_search_mlp = GridSearchCV(mlp_model, param_grid_mlp, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_mlp.fit(X_train_scaled, y_train)
best_mlp = grid_search_mlp.best_estimator_
y_pred_mlp = best_mlp.predict(X_test_scaled)

print("MLP Classifier Results:")
print(classification_report(y_test, y_pred_mlp))
plot_confusion_matrix_heatmap(best_mlp, X_test_scaled, y_test, "MLP")

# Step 6: RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_scaled, y_train)
y_pred_rf = rf_model.predict(X_test_scaled)

print("Random Forest Classifier Results:")
print(classification_report(y_test, y_pred_rf))
plot_confusion_matrix_heatmap(rf_model, X_test_scaled, y_test, "Random Forest")

# Feature Importance for RandomForest
importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1]
plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.bar(range(X_train.shape[1]), importances[indices], align="center")
plt.xticks(range(X_train.shape[1]), [X.columns[i] for i in indices], rotation=90)
plt.show()

# Step 7: SVM with GridSearchCV
param_grid_svm = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['rbf', 'poly', 'sigmoid'],
    'gamma': ['scale', 'auto']
}
svm_model = SVC(random_state=42)
grid_search_svm = GridSearchCV(svm_model, param_grid_svm, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_svm.fit(X_train_scaled, y_train)
best_svm = grid_search_svm.best_estimator_
y_pred_svm = best_svm.predict(X_test_scaled)

print("SVM Classifier Results:")
print(classification_report(y_test, y_pred_svm))
plot_confusion_matrix_heatmap(best_svm, X_test_scaled, y_test, "SVM")

# Step 8: XGBoost Classifier with GridSearchCV
param_grid_xgb = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [4, 6, 8],
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9],
    'gamma': [0, 0.1, 0.3],
    'reg_alpha': [0, 0.01, 0.1],
    'reg_lambda': [1, 1.5, 2]
}

xgb_model = xgb.XGBClassifier(random_state=42, use_label_encoder=False)
grid_search_xgb = GridSearchCV(xgb_model, param_grid_xgb, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_xgb.fit(X_train_scaled, y_train)
best_xgb = grid_search_xgb.best_estimator_
y_pred_xgb = best_xgb.predict(X_test_scaled)

print("XGBoost Classifier Results:")
print(classification_report(y_test, y_pred_xgb))
plot_confusion_matrix_heatmap(best_xgb, X_test_scaled, y_test, "XGBoost")

# Step 9: Logistic Regression
logreg_model = LogisticRegression(random_state=42, max_iter=1000)
logreg_model.fit(X_train_scaled, y_train)
y_pred_logreg = logreg_model.predict(X_test_scaled)

print("Logistic Regression Results:")
print(classification_report(y_test, y_pred_logreg))
plot_confusion_matrix_heatmap(logreg_model, X_test_scaled, y_test, "Logistic Regression")

# Step 10: Gradient Boosting Classifier
gbc_model = GradientBoostingClassifier(random_state=42)
gbc_model.fit(X_train_scaled, y_train)
y_pred_gbc = gbc_model.predict(X_test_scaled)

print("Gradient Boosting Classifier Results:")
print(classification_report(y_test, y_pred_gbc))
plot_confusion_matrix_heatmap(gbc_model, X_test_scaled, y_test, "Gradient Boosting")

# Step 11: AdaBoost Classifier
ada_model = AdaBoostClassifier(random_state=42)
ada_model.fit(X_train_scaled, y_train)
y_pred_ada = ada_model.predict(X_test_scaled)

print("AdaBoost Classifier Results:")
print(classification_report(y_test, y_pred_ada))
plot_confusion_matrix_heatmap(ada_model, X_test_scaled, y_test, "AdaBoost")

# Step 12: Cross-validation for all models
def cross_validate_model(model, X_train, y_train, name):
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    print(f"{name} Cross-validation scores: {cv_scores}")
    print(f"Mean cross-validation score for {name}: {np.mean(cv_scores)}")

# Cross-validate each model
cross_validate_model(best_mlp, X_train_scaled, y_train, "MLP")
cross_validate_model(rf_model, X_train_scaled, y_train, "Random Forest")
cross_validate_model(best_svm, X_train_scaled, y_train, "SVM")
cross_validate_model(best_xgb, X_train_scaled, y_train, "XGBoost")
cross_validate_model(logreg_model, X_train_scaled, y_train, "Logistic Regression")
cross_validate_model(gbc_model, X_train_scaled, y_train, "Gradient Boosting")
cross_validate_model(ada_model, X_train_scaled, y_train, "AdaBoost")

# Step 13: Additional Analytical Visualization

# Plot the distribution of the target variable (Risk) after balancing
plt.figure(figsize=(8, 6))
sns.countplot(x=y_balanced)
plt.title("Distribution of Risk (High/Low) after Oversampling")
plt.show()

# ESG Score Trends over Time
plt.figure(figsize=(12, 6))
for company in esg_cols:
    sns.lineplot(x='Date', y=company, data=data, label=company)

plt.title('ESG Scores Over Time for Each Company')
plt.xlabel('Date')
plt.ylabel('ESG Score')
plt.legend(title='Company')
plt.show()

# Volatility Trends over Time
plt.figure(figsize=(12, 6))
for company in esg_cols:
    sns.lineplot(x='Date', y=f'{company}_Volatility', data=data, label=f'{company} Volatility')

plt.title('Rolling 30-Day Volatility of ESG Scores')
plt.xlabel('Date')
plt.ylabel('Volatility (Standard Deviation)')
plt.legend(title='Company')
plt.show()

# Heatmap of Correlation Matrix
corr_matrix = data[esg_cols + macro_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap: ESG Scores and Macroeconomic Indicators')
plt.show()

# Boxplot of ESG Scores by Company
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[esg_cols])
plt.title("Boxplot of ESG Scores for Different Companies")
plt.ylabel("ESG Score")
plt.xlabel("Company")
plt.show()

# Pairplot to visualize relationships
sns.pairplot(data[esg_cols + macro_cols], diag_kind="kde")
plt.suptitle("Pairplot of ESG Scores and Macroeconomic Data", y=1.02)
plt.show()
```
